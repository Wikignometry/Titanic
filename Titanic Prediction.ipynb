```python
import pandas as pd
import numpy as np

titanic_df = pd.read_csv('tt-train.csv')
titanic_df.head()
```

```python
titanic_df.isnull().sum()
```

```python
# Fill missing values

titanic_df['Age'].fillna(titanic_df['Age'].median(skipna=True), inplace=True)
titanic_df['Embarked'].fillna('S', inplace=True)
```

```python
# Create new variables

titanic_df['Family'] = titanic_df['SibSp'] + titanic_df['Parch']
titanic_df['Minor'] = np.where(titanic_df['Age'] <= 16, 1, 0)
titanic_df = pd.get_dummies(titanic_df, columns=['Pclass','Embarked','Sex'])
```

```python
titanic_df['Master'] = np.where(titanic_df['Name'].str.contains('Master'), 1, 0)
titanic_df['Miss'] = np.where(titanic_df['Name'].str.contains('Miss'), 1, 0)
```

```python
titanic_df.head()
```

```python
titanic_df[np.isnan(titanic_df.Age)].head(20)
```

```python
master_avg = titanic_df[titanic_df.Master == 1]['Age'].median(skipna=True)
print(master_avg)
```

```python
titanic_df['Age'] = np.where(titanic_df['Age'].isnull() & titanic_df['Master'] == 1, master_avg, titanic_df['Age'])
titanic_df[np.isnan(titanic_df.Age)].head(20)
```

```python
miss_avg = titanic_df[titanic_df.Master == 1]['Age'].median(skipna=True)

titanic_df['Age'] = np.where(titanic_df['Age'].isnull() & titanic_df['Miss'] == 1, miss_avg, titanic_df['Age'])
titanic_df[np.isnan(titanic_df.Age)].head(20)
```

```python
titanic_df.head(20)
```

```python
from sklearn.linear_model import LogisticRegression

cols = ['Sex_male','Pclass_1','Pclass_2','Embarked_C','Family','Minor', 'Age'] 
X = titanic_df[cols]
y = titanic_df['Survived']

model = LogisticRegression().fit(X, y)
```

```python
test_df = pd.read_csv('tt-test.csv')
test_df.head()
```

```python
test_df.isnull().sum()
```

```python
test_df['Age'].fillna(test_df['Age'].median(skipna=True), inplace=True)
test_df['Embarked'].fillna('S', inplace=True)
test_df['Family'] = test_df['SibSp'] + test_df['Parch']
test_df['Minor'] = np.where(test_df['Age'] <= 16, 1, 0)
test_df = pd.get_dummies(test_df, columns=['Pclass','Embarked','Sex'])
```

```python
T = test_df[cols]

test_df['Predicted'] = model.predict(T)
test_df['Correct'] = np.where(test_df['Survived'] == test_df['Predicted'], 1, 0)
```
